```python
''' problem statement 3 : one set contains students in the cricket club another in footblall club .
  find the students who are : 1. In both clubs
                              2. In only one club
  concept used : intersectiopn , symetricv diffrence'''



cricket = {"Amit", "Riya", "Sohan", "Neha"}
football = {"Riya", "Neha", "Karan", "Priya"}

both_clubs = cricket & football

only_one_club = cricket ^ football


print("Students in both clubs:", both_clubs)
print("Students in only one club:", only_one_club)


```
